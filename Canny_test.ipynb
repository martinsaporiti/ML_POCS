```python
# Canny Edge Detection is a popular edge detection algorithm. It was developed by John F. Canny in
# * It is a multi-stage algorithm and we will go through each stages.
# * Noise Reduction
# Since edge detection is susceptible to noise in the image, first step is to remove 
# the noise in the image with a 5x5 Gaussian filter. We have already seen this in previous chapters.
```

```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
img = cv.imread('images/abba.png',0)
edges = cv.Canny(img,100,200)
plt.subplot(121),plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
plt.show()
```
